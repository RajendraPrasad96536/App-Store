```python
import cv2
import numpy as np
import random
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/photo.jpg')
```

```python
image = cv2.imread('/content/photo.jpg',cv2.IMREAD_GRAYSCALE)
mean = 0
var = 50
for i in range(0, 3):
  sigma = var ** i
  gaussian_noise = np.random.normal(mean, sigma, (image.shape[0],image.shape[1])) 

  noisy_image1 = image+gaussian_noise

  cv2_imshow(noisy_image1)
```

```python
uniform_noise = np.zeros((image.shape[0], image.shape[1]),dtype=np.uint8)
#print(uniform_noise)
cv2.randu(uniform_noise,0,255)
for i in [0.5,0.8, 1, 1.5, 2, 2.5, 5]:
  uniform_noise = (uniform_noise*0.5).astype(np.uint8)
  #print(uniform_noise)
  noisy_image2 = cv2.add(image, uniform_noise)

  cv2_imshow(noisy_image2)
```
